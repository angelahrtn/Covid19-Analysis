```python
# Melihat kolom yang tersedia
raw_data_confirmed.keys()

# Menghapus kolom yang tidak diperlukan (Province/State, Lat, Long)
confirm_data = raw_data_confirmed

# Mengubah data dari wide format ke long format dengan memindahkan kolom tanggal menjadi satu kolom 'Date'
confirm_data_melted = confirm_data.melt(id_vars=['Country/Region'],
                                        var_name='Date',
                                        value_name='Total Cases')

# Mengubah kolom 'Date' menjadi format datetime
confirm_data_melted['Date'] = pd.to_datetime(confirm_data_melted['Date'])

# Mengelompokkan data berdasarkan 'Country/Region' dan 'Date', serta menjumlahkan kasus
confirm_data_grouped = confirm_data_melted.groupby(['Country/Region', 'Date']).sum().reset_index()

# Menambahkan kolom bulan untuk pengelompokan
confirm_data_grouped['Month'] = confirm_data_grouped['Date'].dt.to_period('M')

# Mengelompokkan data berdasarkan 'Country/Region' dan 'Month', serta menjumlahkan total kasus per bulan
monthly_confirm_data = confirm_data_grouped.groupby(['Country/Region', 'Month']).agg({'Total Cases': 'sum'}).reset_index()

# Mengubah kembali 'Month' menjadi datetime untuk kemudahan plotting
monthly_confirm_data['Month'] = monthly_confirm_data['Month'].dt.to_timestamp()

# Menentukan tanggal akhir untuk grafik
end_date = pd.Timestamp('2020-04-07')

# Menyaring data hingga tanggal akhir yang ditentukan
monthly_confirm_data = monthly_confirm_data[monthly_confirm_data['Month'] <= end_date]

# Mengatur ukuran plot
plt.figure(figsize=(14, 7))

# Membuat line plot dengan seaborn berdasarkan bulan
sns.lineplot(data=monthly_confirm_data,
             x='Month',
             y='Total Cases',
             hue='Country/Region',
             ci=None)

# Memberikan judul dan label sumbu
plt.title('Perkembangan Total Kasus COVID-19 per Bulan di Berbagai Negara (Hingga 7 April 2020)', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Total Kasus', fontsize=12)

# Menambahkan grid untuk membantu visualisasi
plt.grid()

# Memutar label sumbu-x agar lebih mudah dibaca
plt.xticks(rotation=45)

# Menyembunyikan legend dari grafik
plt.legend(title='Country/Region', bbox_to_anchor=(1.05, 1), loc='upper left', frameon=False)

# Menampilkan grafik
plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
plt.show()
```

### 4. Penyelarasan kurva pertumbuhan seluruh negara dengan jumlah kasus kematian COVID-19  25 (Bobot: 20%)

**Untuk mendapatkan gambaran perkembangan kasus covid-19 di berbagai negara, diperlukan penyelarasan kurva pertumbuhan setiap negara. Kurva hanya menampilkan informasi yang dimulai dengan hari dimana data kasus kematian COVID-19 setidaknya 25 orang.**

```python
# Menghapus kolom yang tidak diperlukan
raw_data_deaths_cleaned = raw_data_deaths

# Mengubah data dari wide format ke long format
deaths_data_melted = raw_data_deaths_cleaned.melt(id_vars=['Country/Region'],
                                                    var_name='Date',
                                                    value_name='Total Deaths')

# Mengubah kolom 'Date' menjadi format datetime
deaths_data_melted['Date'] = pd.to_datetime(deaths_data_melted['Date'])
```

```python
# Mengelompokkan data berdasarkan negara dan tanggal untuk mendapatkan total kematian
deaths_grouped = deaths_data_melted.groupby(['Country/Region', 'Date']).sum().reset_index()

# Mengfilter negara dengan total kematian  25
countries_with_deaths = deaths_grouped[deaths_grouped['Total Deaths'] >= 25]

# Menentukan tanggal pertama dimana total kematian mencapai 25 untuk setiap negara
first_date = countries_with_deaths.groupby('Country/Region')['Date'].min().reset_index()
first_date.rename(columns={'Date': 'Start Date'}, inplace=True)

# Menggabungkan kembali data untuk mendapatkan informasi yang lengkap
merged_data = pd.merge(deaths_grouped, first_date, on='Country/Region')

# Menghitung jumlah kematian dari tanggal pertama
merged_data['Days Since 25 Deaths'] = (merged_data['Date'] - merged_data['Start Date']).dt.days

# Mengfilter untuk hanya menyertakan hari setelah mencapai 25 kematian
final_data = merged_data[merged_data['Days Since 25 Deaths'] >= 0]

# Visualisasi
plt.figure(figsize=(15, 8))
sns.lineplot(data=final_data, x='Days Since 25 Deaths', y='Total Deaths', hue='Country/Region', linewidth=2.5)

# Memberikan judul dan label
plt.title('Penyelarasan Kurva Pertumbuhan Kasus Kematian COVID-19 di Negara dengan Kematian  25', fontsize=16, fontweight='bold')
plt.xlabel('Hari Sejak 25 Kematian', fontsize=12)
plt.ylabel('Total Kematian', fontsize=12)

# Mengatur legend
plt.legend(title='Negara', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Menambahkan grid dengan ketebalan garis
plt.grid(color='grey', linestyle='--', linewidth=0.5)

# Menyesuaikan tampilan
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Menampilkan grafik
plt.tight_layout()  # Untuk memastikan layout tidak terpotong
plt.show()
```

Grafik menunjukkan perkembangan kematian COVID-19 dari hari ke hari sejak titik di mana jumlah kematian mencapai 25 untuk setiap negara.

Visualisasi ini memberikan wawasan yang jelas tentang bagaimana negara-negara yang terpengaruh COVID-19 dengan jumlah kematian yang signifikan ( 25) mengalami perkembangan kasus kematian seiring waktu. Dengan menyelaraskan kurva dari titik di mana kematian mencapai 25, kita dapat dengan mudah membandingkan tren antara negara yang berbeda. Ini memungkinkan analisis yang lebih mendalam mengenai efektivitas langkah-langkah kesehatan masyarakat dan respons terhadap pandemi di berbagai belahan dunia. ### 5. Visualisasi jumlah pasien yang sembuh (recovered people)  (Bobot: 25%)
    
**Anda perlu mendapatkan informasi mengenai total pasien COVID-19 yang dinyatakan sembuh seperti pada kasus terkonfirmasi dan kasus kematian tapi, kali ini lakukan hal tersebut dengan membuat `user-defined functions` yang diperlukan bukan mengetik ulang barisan code yang sama seperti di nomor sebelumnya.**

```python
# Fungsi untuk membersihkan dan mengolah data
def process_recovered_data(data):
    # Menghapus kolom yang tidak diperlukan
    data_cleaned = data.drop(columns=['Province/State', 'Lat', 'Long'])

    # Mengubah data dari wide format ke long format
    data_melted = data_cleaned.melt(id_vars=['Country/Region'],
                                      var_name='Date',
                                      value_name='Total Recovered')

    # Mengubah kolom 'Date' menjadi format datetime
    data_melted['Date'] = pd.to_datetime(data_melted['Date'])

    # Mengelompokkan data berdasarkan negara dan tanggal untuk mendapatkan total sembuh
    recovered_grouped = data_melted.groupby(['Country/Region', 'Date']).sum().reset_index()

    return recovered_grouped
```

```python
# Mengolah data
recovered_data = process_recovered_data(raw_data_recovered)

# Visualisasi jumlah pasien yang sembuh
plt.figure(figsize=(15, 8))

# Menggunakan palet warna yang menarik dan mengatur ketebalan garis
sns.lineplot(data=recovered_data, x='Date', y='Total Recovered', hue='Country/Region', linewidth=2.5)

# Memberikan judul dan label
plt.title('Total Pasien COVID-19 yang Sembuh per Negara', fontsize=16, fontweight='bold')
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Total Pasien Sembuh', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Negara', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menambahkan grid
plt.grid()
```

**Lakukan visualisasi yang sama pada beberapa negara berikut (France, Spain, China, US, Italy, and Australia). Berikan judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
# Fungsi untuk membersihkan dan mengolah data
def process_recovered_data(data, selected_countries):
    # Menghapus kolom yang tidak diperlukan
    data_cleaned = data.drop(columns=['Province/State', 'Lat', 'Long'])

    # Mengubah data dari wide format ke long format
    data_melted = data_cleaned.melt(id_vars=['Country/Region'],
                                      var_name='Date',
                                      value_name='Total Recovered')

    # Mengubah kolom 'Date' menjadi format datetime
    data_melted['Date'] = pd.to_datetime(data_melted['Date'])

    # Mengelompokkan data berdasarkan negara dan tanggal untuk mendapatkan total sembuh
    recovered_grouped = data_melted.groupby(['Country/Region', 'Date']).sum().reset_index()

    # Memfilter data untuk negara yang dipilih
    filtered_data = recovered_grouped[recovered_grouped['Country/Region'].isin(selected_countries)]

    return filtered_data
```

```python
# Daftar negara yang akan divisualisasikan
selected_countries = ['France', 'Spain', 'China', 'US', 'Italy', 'Australia']

# Mengolah data
recovered_data = process_recovered_data(raw_data_recovered, selected_countries)

# Visualisasi jumlah pasien yang sembuh untuk negara terpilih
plt.figure(figsize=(15, 8))

# Menggunakan palet warna yang menarik dan mengatur ketebalan garis
sns.lineplot(data=recovered_data, x='Date', y='Total Recovered', hue='Country/Region',
             palette='tab10', linewidth=2.5)

# Memberikan judul dan label
plt.title('Total Pasien COVID-19 yang Sembuh di Negara Terpilih', fontsize=16, fontweight='bold')
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Total Pasien Sembuh', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Negara', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menambahkan grid
plt.grid()

# Menampilkan grafik
plt.tight_layout()  # Untuk mengatur layout agar lebih rapi
plt.show()
```
